# 1. 平稳/非平稳  
&emsp; &emsp; 一个例子，创建两个序列$A$ 和 $B$. 序列 $A$ 是由固定参数生成的，即不随时间而改变，序列 $B$ 是随着时间的改变而改变的。

```python
def generate_datapoint(params):
    mu = params[0]
    sigma = params[1]
    return np.random.normal(mu, sigma)
```

### 序列 $A$

```python
# Set the parameters and the number of datapoints
params = (0, 1)
T = 100

A = pd.Series(index=range(T))
A.name = 'A'

for t in range(T):
    A[t] = generate_datapoint(params)
plt.figure(figsize=(16,8))
plt.plot(A)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series A']);
```

### 序列 $B$

```python
# Set the number of datapoints
T = 100

B = pd.Series(index=range(T))
B.name = 'B'

for t in range(T):
    # Now the parameters are dependent on time
    # Specifically, the mean of the series changes over time
    params = (t * 0.1, 1)
    B[t] = generate_datapoint(params)
plt.figure(figsize=(16,8))
plt.plot(B)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series B']);
```

### 为什么说非平稳是危险的

&emsp; &emsp; 很多统计学的检验，都要求被检验数据是平稳的。如果在非平稳的数据集上做检验，会得到无用的结果。例如，在非平稳序列$B$上来测试下。

```python
# m 是 b的均值
m = np.mean(B)
plt.figure(figsize=(16,8))
plt.plot(B)
plt.hlines(m, 0, len(B), linestyles='dashed', colors='r')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series B', 'Mean'])
```

&emsp; &emsp; 这个均值对未来的预测没有什么用处。

### 数据的平稳检验

```python
def check_for_stationarity(X, cutoff=0.01):
    # H_0 in adfuller is unit root exists (non-stationary)
    # We must observe significant p-value to convince ourselves that the series is stationary
    # 函数adfuller()是做平稳性检验的。Augmented Dickey-Fuller unit root test
    pvalue = adfuller(X)[1]  # p值
    if pvalue < cutoff:
        print 'p-value = ' + str(pvalue) + ' 序列 ' + X.name +' 是平稳的.'
        return True
    else:
        print 'p-value = ' + str(pvalue) + ' 序列 ' + X.name +' 是非平稳的.'
        return False
```

```python
check_for_stationarity(A);
check_for_stationarity(B);
```

&emsp; &emsp;可以确定的是，序列均值的改变会使其非平稳。但是下面这个例子会有些不同。

```python
# Set the number of datapoints
T = 100

C = pd.Series(index=range(T))
C.name = 'C'

for t in range(T):
    # Now the parameters are dependent on time
    # Specifically, the mean of the series changes over time
    params = (np.sin(t), 1)    # 均值为时间的正弦函数
    C[t] = generate_datapoint(params)

plt.figure(figsize=(16,8))
plt.plot(C)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series C']);
```

&emsp; &emsp;所以均值的来回移动很难排除是噪声的影响。

```python
check_for_stationarity(C);
```

# 2. 单整阶数

### 移动平均表达（世界定理） 

时间序列 $Y_t$ 表示为： 

$$Y_t = \sum_{j=0}^\infty b_j \epsilon_{t-j} + \eta_t$$

* $\epsilon$ 是新息过程('innovation' series)
* $b_j$ 是新息过程的移动平均权重
* $\eta$ 是确定的序列

关键在于 $\eta$ 是确定的，比如正弦波，因此我们可以很好的拟合它。新息过程是随机的，随着时间推进会有新的信息产生。特别的， $\epsilon_t = \hat Y_t - Y_t$， $\hat Y_t$是 $Y_t$ 的最优预测（仅仅用 $t$ 之前的信息）。每个 $b_j$ 只能够说明 $\epsilon$ 对预测 $Y_t$ 的影响度。


### 单整

$i$ 阶单整记为：$I(i)$

当如下条件满足时，一个时间序列被称为 $I(0)$ 。

$$\sum_{k=0}^\infty |b_k|^2 < \infty$$
#### 引用：
https://en.wikipedia.org/wiki/Order_of_integration
https://en.wikipedia.org/wiki/Wold%27s_theorem

####  $I(0)$检验

&emsp; &emsp; 在实际检验中，自相关系数的和是否为有限也许不能确定。但是如果我们有了一个有限的数据集和有限的相关系数，这时和就是有限的。“Given this difficulty, tests for $I(0)$ rely on stationarity implying the property.” 如果我们发现时间序列时平稳的，那么一定是 $I(0)$。

&emsp; &emsp; 继续使用序列 $A$ . 因为 $A$ 是平稳的，所以也是 $I(0)$。

```python
plt.figure(figsize=(16,8))
plt.plot(A)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series A']);
```

### 单整的归纳

&emsp; &emsp; 如果对一个 $I(0)$ 序列累积求和,新的序列将是 $I(1)$.类似地，为了得到 $I(n)$ 对一个 $I(0)$ 序列做迭代，求 $n$ 次累积和。

&emsp; &emsp; 现在来创建一个 $I(1)$ 序列和 $I(2)$ 序列（通过对 $A$ 累积求和）

```python
A1 = np.cumsum(A)
plt.figure(figsize=(16,8))
plt.plot(A1)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series A1']);
```

```python
A2 = np.cumsum(A1)
plt.figure(figsize=(16,8))
plt.plot(A2)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend(['Series A2']);
```

### “ Breaking Down Orders of Integration ” 

&emsp; &emsp; 相反地, 为了找到一个给定序列的单整，我们要做累计求和的逆操作（差分）$\Delta$ 。 特别地，

$$(1-L) X_t = X_t - X_{t-1} = \Delta X$$

$$(1-L)^d X_t$$

$L$ 是滞后算子，有时候记为 $B$ for 'backshift'。  

$$L X_t = X_{t-1}$$

and

$$(1-L) X_t = X_t - X_{t-1}$$

一个序列 $Y_t$ 是 $I(1)$ 如果  $Y_t - Y_t-1$ 是 $I(0)$. 换言之，如果对一个 $I(0)$ 序列累计求和, 会得到一个 $I(1)$ 序列。

## 重要提示

&emsp; &emsp; 我们不能因为一个时间序列过去是平稳的，就认为未来也会是平稳的。平稳的一致性检验，比如交叉检验和样本外检验都是很有必要的。这是适合任何一种统计特性的。


### 收益率分析

&emsp; &emsp; 由于收益率比价格序列更平稳，因此在量化金融中更多的被用到，这会更容易建立模型。预测价格相对更加困难，而用一个预测收益率的模型去预测价格是不可取地(tricky),因为这会让收益率预测时的错误随着时间被逐渐放大。

## 3. 协整

最后, 我们来讨论协整。

### Def: 线性组合

一个时间序列的线性组合 ($X_1$, $X_2$, $\dots$, $X_k$) 是一个新的时间序列 $Y$ ，由以下公式得出，对任何实数集 $b_1 \dots b_k$

$$Y = b_1X_1 + b_2X_2 + \dots + b_kX_k$$

### 正式定义
协整的正式定义：
对于一些时间序列 ($X_1$, $X_2$, $\dots$, $X_k$)，如果所有序列都是 $I(1)$,并且一些序列是 $I(0)$，就称为协整的。
###  举例

$X_1$, $X_2$, 和 $X_3$ 都是 $I(1)$, 并且 $2X_1 + X_2 + 0X_3 = 2X_1 + X_2$ 是 $I(0)$. 此时，这些时间序列就是协整的。

### 模拟数据举例

```python
# Length of series
N = 100

# Generate a stationary random X1
X1 = np.random.normal(0, 1, N)
# Integrate it to make it I(1)
X1 = np.cumsum(X1)
X1 = pd.Series(X1)
X1.name = 'X1'

# Make an X2 that is X1 plus some noise
X2 = X1 + np.random.normal(0, 1, N)
X2.name = 'X2'
```

```python
plt.figure(figsize=(16,8))
plt.plot(X1)
plt.plot(X2)
plt.xlabel('Time')
plt.ylabel('Series Value')
plt.legend([X1.name, X2.name]);
```

因为 $X_2$ 是 $I(1)$ 序列加上固定的噪声，他应该还是 $I(1)$ 。

```python
Z = X2.diff()[1:]
Z.name = 'Z'

check_for_stationarity(Z);
```

现在来展示协整，我们要找到 $X_1$ 和 $X_2$ 的线性组合是平稳的。比如 $X_2-X_1$。

```python
Z = X2 - X1
Z.name = 'Z'

plt.figure(figsize=(16,8))
plt.plot(Z)
plt.xlabel('Time')
plt.ylabel('Series Value')
plt.legend(['Z']);

check_for_stationarity(Z);
```

### 协整的检验

有一些方法来检验协整性。 这里 [wikipedia article](https://en.wikipedia.org/wiki/Cointegration) 提到了一些。 我们只需要解出生成一个 $I(0)$ 线性组合的这些系数 $b_1, \dots b_k$ 。如果我们最优的猜测没有通过统计检验，我们会拒绝协整的假设。 
实际上，一个常见的方法是对这一对时间序列做线性回归去去估计下面模型的 $\beta$ 。

$$X_2 = \alpha + \beta X_1 + \epsilon$$

### 第三方函数

有第三方的开源库提供了这个函数。[documentation](http://statsmodels.sourceforge.net/devel/_modules/statsmodels/tsa/stattools.html)

```python
from statsmodels.tsa.stattools import coint

coint(X1, X2)
```
