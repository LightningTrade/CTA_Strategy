```python
import time
import numpy as np
from math import pi
import pandas as pd
from bokeh.io import output_notebook
from bokeh.plotting import figure, show, output_file,ColumnDataSource
from bokeh.models import Rect, HoverTool, BoxZoomTool,HelpTool,Range1d, LinearAxis, WheelZoomTool, PanTool, ResetTool, ResizeTool,SaveTool
output_notebook()
%matplotlib inline
```

加载本地的数据

```python
from talib import MA
quotes = pd.read_csv('../../ts_data/J00.DCE.csv',index_col=0)
```

```python
quotes= quotes.dropna()
openp=quotes['open']
closep=quotes['close']
highp=quotes['high']
lowp=quotes['low']
volume=quotes['vol']
time=pd.to_datetime(quotes.index)
#date=[x.strftime("%Y-%m-%d") for x in quotes.index]
quotes['date']=pd.to_datetime(quotes.index)

w = 12*60*60*1000 # half day in ms
mids = (openp + closep)/2
spans = abs(closep-openp)
inc = closep > openp
dec = openp > closep 

x_rect_src_inc =ColumnDataSource(
            data=dict(
                    date=quotes['date'],
                    open=openp,
                    close=closep,
                    high=highp,
                    low=lowp))
'''
## TODO:未能正确显示数据
hover = HoverTool(
    tooltips=[("index","$index"),
            ("date", "@date"),
            ("open", "@open"),
            ("close", "@close"),
            ("high", "@high"),
            ("low", "@low"),])
'''
TOOLS = [HoverTool(), WheelZoomTool(), ResizeTool(), BoxZoomTool(),
         ResetTool(),PanTool(), SaveTool()]

p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=1000)
p.xaxis.major_label_orientation = pi/4
p.grid.grid_line_alpha=0.3
p.background_fill_color = "black"
#time[inc], mids[inc], w, spans[inc]
#p.segment(time, highp, time, lowp, color="black")
p.rect(time[inc], mids[inc], w, spans[inc], fill_color="red", line_color="red")
p.rect(time[dec], mids[dec], w, spans[dec], fill_color="green", line_color="green")

p.segment(time[inc], highp[inc], time[inc], lowp[inc], color="red")
p.segment(time[dec], highp[dec], time[dec], lowp[dec], color="green")

'''划线，以均线为例子'''

MAs = [ 5, 10, 15, 20 ]
line_color = [ 'white', 'yellow', 'red', 'green' ]
for ma_th, ma in enumerate( MAs ):
        legend = 'MA' + str( ma )  
        p.line( x = time, y = MA( quotes['close'].as_matrix(), ma, matype=1 ), 
                   legend = legend, line_color = line_color[ ma_th ] )
#output_file("candlestick.html", title="candlestick.py example")
show(p)  # open a browser
```
