```python
import pprint
import numpy as np
import scipy.linalg   # SciPy Linear Algebra Library

A = np.array([ [7, 3, -1, 2], [3, 8, 1, -4], [-1, 1, 4, -1], [2, -4, -1, 6] ])
P, L, U = scipy.linalg.lu(A)
lu, piv = scipy.linalg.lu_factor(A)

print "A:"
pprint.pprint(A)

print "P:"
pprint.pprint(P)

print "L:"
pprint.pprint(L)

print "U:"
pprint.pprint(U)

print "lu"
pprint.pprint(lu)

print "piv"
pprint.pprint(piv)
```

```python
B = np.array([[1],[2],[3],[4]])
X = scipy.linalg.lu_solve((lu,piv), B)
```
